```python
import pandas as pd
```

```python
df = pd.read_csv("Customers.csv")
```

1) Insert a new column called Initials that combine the first letter of First name and Last
name (use .map() a Lambda function). For example: Joseph Martins -> JM

```python
df['Initials'] = df['FirstName'].map(lambda s: s[0]) + df['LastName'].map(lambda s: s[0])
df[['FirstName', 'LastName', 'Gender', 'Initials']]
```

2) Find the number of duplicates and remove the duplicated, and only keep the first entry

```python
print("Number of duplicate:", df.duplicated(keep='first').sum())
df.drop_duplicates(keep='first')
```

3) Find the number of professions

```python
print("Number of professions:", df['Profession'].nunique())
```

4) Find the average income

```python
print(df['Annual Income ($)'].mean())
```

5) Find the median income

```python
print(df['Annual Income ($)'].median())
```

6) Extract info [FirstName, LastName, Gender, Age, Annual Income, Spending Score,
Profession] of Male engineers above 20 years old with income over 60000 $

```python
df = df[['FirstName', 'LastName', 'Gender', 'Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Profession']]
df[df['Annual Income ($)'] >= 60000]
```

7) Extract info [FirstName, LastName, Gender, Age, Profession] of Female ar sts between
20-40 years old with income below 40000 $

```python
df[['FirstName', 'LastName', 'Gender', 'Age', 'Profession']]
df[(df['Gender'] == 'Female') & (df['Annual Income ($)'] < 40000)]
```

8) Extract those whose first name is Jac (use Series.str.contains())

```python
df[df['FirstName'].str.contains('Jac')]
```

9) Extract those whose spending score is above 70

Extract males whose spending score is above 70.

Extract females whose spending score is above 70.

```python
df[df['Spending Score (1-100)'] >= 70]
```

```python
df[(df['Gender'] == 'Male') & (df['Spending Score (1-100)'] >= 70)]
```

10) Find the percentage of Male with spending score above 70 vs percentage of Female with spending score above 70

```python
male_score = df[(df['Gender'] == 'Male') & (df['Spending Score (1-100)'] >= 70)]
female_score = df[(df['Gender'] == 'Female') & (df['Spending Score (1-100)'] >= 70)]
print("Percentage of Male with spending score above 70:", len(male_score) / len(df) * 100, "%")
print("Percentage of Female with spending score above 70:", len(female_score) / len(df) * 100, "%")
```
