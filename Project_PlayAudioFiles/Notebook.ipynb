## This exercise play and write an audio file, then draw a sine wave

Import the required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import librosa  # Install this with pip install librosa
import librosa.display
from pydub import AudioSegment  # Install this with pip install pydub
from pydub.playback import play
import os
```

Load an audio file and then convert it from mp3 to wav
Note: We must download ffmpeg and ffprobe for this to work. Simply write: brew install ffmpeg and brew install ffprobe

```python
file_path = 'birdsinging.mp3'
audio = AudioSegment.from_mp3(file_path)
audio.export("birdsinging.wav", format="wav")
print("Converted mp3 to wav!!")
# Play the audio file
print("Playing the audio file...", file_path)
play(audio)
```

Write the audio file with librosa, return the total samples (length of y), and sr (the sample rate). Duration in seconds is len(y)/sr

```python
y, sr = librosa.load("birdsinging.wav", sr=None)
print(f"Sample Rate: {sr}, \nAudio Data Shape: {len(y)}")
print(f"Bits per sample: 16")
print(f"Duration: {len(y)/sr} seconds")
```

Get the mb size of the audio file, and mb per second

```python
file_size = os.path.getsize("birdsinging.wav") / (1024 * 1024)
print(f"File Size: {file_size:.2f} MB")
print(f"File Size per Second: {file_size / (len(y)/sr):.2f} MB/s")
```

Get the Max and Min Amplitude

```python
print(f"Max Amplitude: {max(y)}")
print(f"Min Amplitude: {min(y)}")
```

Plot a waveform

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
